```python
import re
from collections import defaultdict

def token(message):
    message =  message.lower()
    all_words = re.findall("[a-z0-9]+",message)
    return set(all_words)

def count_words(tr_set):
    counts = defaultdict(lambda :[0,0])
    for message , spam in tr_set:
        for word in token(message):
            counts[word][0 if spam else 1] += 1
    return counts

def word_prob(counts,total_spam,total_no_spam,k=0.5):
    return [(w,
             (spam+k)/(total_spam + 2*k),
             (no_spam+k)/(total_no_spam + 2*k))
            for w ,(spam,no_spam) in counts.iteritems()]
import math

def spam_prob(word_prob,message):
    message_word = token(message)
    log_p_spam = log_p_no_spam = 0
    for word ,p_spam,p_no_spam = word_prob:
        if word in message_word:
            log_p_spam += math.log(p_spam)
            log_p_no_spam += math.log(p_no_spam)
        else:
            log_p_spam += math.log(1-p_spam)
            log_p_no_spam += math.log(1-p_no_spam)
        
        p_spam = math.exp(log_p_spam)
        p_no_spam = math.exp(log_p_no_spam)
        return p_spam / (p_spam + p_no_spam)
    
class NaiveBays:
    def ___init__(self,k=0.5):
        self.k = k
        self.word_p = []
        
    def train(self ,tr_set):
        n_spam = len([is_spam for message , is_spam in tr_set if is_spam])
        n_n_spam = len(tr_set)- n_spam
        word_counts = count_words(tr_set)
        self.word_p = word_prob(word_counts,n_spam,n_n_spam,self.k)
    
    def classify(self,message):
        return spam_prob(self.word_p,message)
    
import glob,re
path = "" #file path
data = []

for fn in glob.glob(path):
    is_spam = "ham" not in fn
    
    with open(fn ,"r") as file:
        for line in file:
            if line.startwith("Subject"):
                subject = re.sub(r"Subject:","",line).strip()
                data.append((subject,is_spam))
                
random.seed(0)
train_data,test_data = split_data(data,0.75)
classifier = NaiveBayes()
classifier.train(train_data)
classified = [(subject,is_spam,classifier.classify(subject)) for subject ,is_spam in test_data]
counts = Counter((is_spam,spam_p > 0.5) for _, is_spam,spam_p in classified)
```
